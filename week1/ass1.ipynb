```python
model = MeanShift(bandwidth=0.1)
predict = model.fit_predict(X)
```

```python
labels = model.labels_
print(labels)
uniques = np.unique(labels)
n_clusters = len(uniques)
print('number of clusters: {}'.format(n_clusters))
```

```python
top_labels = [l for l in labels if np.count_nonzero(labels == l) > 15]
```

```python
print(len(np.unique(top_labels)))
```

```python
centers = model.cluster_centers_
```

```python
top_centers = centers[np.unique(top_labels)]
```

```python
len(top_centers)
```

```python
carn = list()
carn.append([33.751277, -118.188740])
carn.append([25.867736, -80.324116])
carn.append([51.503016, -0.075479])
carn.append([52.378894, 4.885084])
carn.append([39.366487, 117.036146])
carn.append([-33.868457, 151.205134])
```

```python
carn
```

```python
dists = list()
for i in range(len(top_centers)):
    for j in range(len(carn)):
        dist = np.sqrt(np.square(top_centers[i][0] - carn[j][0]) + np.square(top_centers[i][1] - carn[j][1]))
        dists.append(dist)
        print("center # {}, - office # {}, dist: {}".format(i, j, dist))
```
